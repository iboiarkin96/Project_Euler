```python
%%time
def polindrom(start_number, end_number):
    list_of_mult = []
    for i in range(start_number, end_number):
        for j in range(start_number, end_number):
            if i != j:
                a = i * j 
                if list(str(a))==list(str(a))[::-1]:
                    list_of_mult.append(a)
    return max(list_of_mult)

polindrom(100, 1000)
```

```python
%%time
def compute(start_number, end_number):
    ans = max(i * j for i in range(start_number, end_number) for j in range(start_number, end_number) if str(i * j) == str(i * j)[ : : -1])
    return str(ans)
compute(100, 1000)
```
